# Conclusions

- GRU VS LSTM - both demonstrated strong predictive capabilities when trained on Google's historical stock data. Both captured trends well, but GRU slightly outperformed LSTM in terms of lower RMSE and better adaptability to fluctuations in stock prices. GRU is also computationally less expensive than LSTM. But still both of them run very fast, which makes them a perfect candidate for scenarios with limited computational resources or a need for faster times.
- Prediction based on correlated stock (using PyCaret for predicting google stock prices based on Meta's data (and vice versa)) highlighted the challenge of correlation-based prediction. Predicting Meta from Google showed slightly better performance, with trends being better aligned, though actual values deviated significantly. The key takeaway is that correlated data might not translate to strong predictive power.
- DeepAR was promising tool for time-series forecasting, but the implementation faced significant issues. Prediction intervals were broad and diverged drastically from the test data. After multiple attempts to debug (redefining the dataset, modifying hyperparameters, checking the pipeline), the root cause of the poor performance was not conclusively identified. This underlines the imprortance of robust dataset preparation and debugging pipelines....
- Linear Regression predicted historical data well, but its reliance on linear assumptions limits its ability to capture more complex relationships. The correlation-based predections were also underwhelming as Linear Regression struggled to leverage Meta's data for predicintg Google's prices.
In all of the notebooks that are associated with models there are also insights that were gathered and arose during the creation of the project.
- GRU emerged as the most effective model, balancing accuracy and computational efficiency.
- the LSTM model remains a viable alternative for scenarios requiring more nuanced long-term dependencies. 
- PyCaret provided a framework for rapid prototyping but struggled with cross-stock prediction tasks.
- DeepAR suffered from significant implementation issues, preventing meaningful insights. But it's cloud-based scalability should help.
