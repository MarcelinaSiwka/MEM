```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = 'clean_output.csv'
data = pd.read_csv(file_path)
```

```python
need = ['old/new', 'town/city', 'district', 'county', 'year', 'month',
       'day', 'property_type_is__D', 'property_type_is__F',
       'property_type_is__O', 'property_type_is__S', 'property_type_is__T']
```

```python
lin_df = data.copy()
lin_df.head()
```

```python
lin_df['town/city'] = lin_df['town/city'].factorize()[0].astype('float32')
lin_df['district'] = lin_df['district'].factorize()[0].astype('float32')
lin_df['county'] = lin_df['county'].factorize()[0].astype('float32')
```

```python
from sklearn.model_selection import train_test_split
X = lin_df[need]
y = lin_df['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train.shape, y_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

normalized_array = scaler.fit_transform(X_train[need])
normalized_X_train = pd.DataFrame(normalized_array, columns=need)

normalized_X_train
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split


# Initialize the Gradient Boosting Regressor
gb = GradientBoostingRegressor(random_state=42)

# Train the model on the training set
gb.fit(normalized_X_train, y_train)

# Make predictions on the training and testing sets
y_train_pred = gb.predict(normalized_X_train)
y_test_pred = gb.predict(X_test)

# Evaluate the model's performance
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)

train_rmse = mean_squared_error(y_train, y_train_pred, squared=False)
test_rmse = mean_squared_error(y_test, y_test_pred, squared=False)

# Print the performance metrics
print("Training R^2 Score:", train_r2)
print("Testing R^2 Score:", test_r2)
print("Training RMSE:", train_rmse)
print("Testing RMSE:", test_rmse)
##############################################

```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error

# Initialize and train the Gradient Boosting Regressor
gb = GradientBoostingRegressor(random_state=42)
gb.fit(X_train, y_train)

# Function to visualize model performance
def visualize_model_performance(model, X_train, X_test, y_train, y_test):
    # Make predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Calculate residuals
    residuals_train = y_train - y_train_pred
    residuals_test = y_test - y_test_pred

    # Calculate performance metrics
    train_r2 = r2_score(y_train, y_train_pred)
    test_r2 = r2_score(y_test, y_test_pred)
    train_rmse = mean_squared_error(y_train, y_train_pred, squared=False)
    test_rmse = mean_squared_error(y_test, y_test_pred, squared=False)

    print("Training R^2 Score:", train_r2)
    print("Testing R^2 Score:", test_r2)
    print("Training RMSE:", train_rmse)
    print("Testing RMSE:", test_rmse)

    # Plot actual vs predicted values
    plt.figure(figsize=(12, 5))
    
    # Scatter plot for training set
    plt.subplot(1, 2, 1)
    plt.scatter(y_train, y_train_pred, alpha=0.6, label='Training Data')
    plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], color='red', linestyle='--', label='Perfect Fit')
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.title('Actual vs Predicted (Training)')
    plt.legend()
    
    # Scatter plot for testing set
    plt.subplot(1, 2, 2)
    plt.scatter(y_test, y_test_pred, alpha=0.6, label='Testing Data', color='orange')
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', label='Perfect Fit')
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.title('Actual vs Predicted (Testing)')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

    # Plot residuals
    plt.figure(figsize=(12, 5))
    
    # Residuals for training set
    plt.subplot(1, 2, 1)
    plt.hist(residuals_train, bins=30, alpha=0.7, label='Training Residuals')
    plt.axvline(x=0, color='red', linestyle='--', label='Zero Error')
    plt.xlabel('Residuals')
    plt.ylabel('Frequency')
    plt.title('Residuals (Training)')
    plt.legend()

    # Residuals for testing set
    plt.subplot(1, 2, 2)
    plt.hist(residuals_test, bins=30, alpha=0.7, label='Testing Residuals', color='orange')
    plt.axvline(x=0, color='red', linestyle='--', label='Zero Error')
    plt.xlabel('Residuals')
    plt.ylabel('Frequency')
    plt.title('Residuals (Testing)')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# Example usage with GradientBoostingRegressor
visualize_model_performance(gb, X_train, X_test, y_train, y_test)
```
